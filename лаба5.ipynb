#№ 5.2.6

Дан список из чисел.

Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий делитель)

```python
def gcd(first, second):
  if second > first:
    x = second
  else:
    x = first
  for i in range(1,x+1):
    if (first %  i== 0) and (second % i == 0 ):
      gcd = i
  return gcd


def lcm(first, second):
  if first > second:
    x = first
  else:
    x = second
  while(True):
    if (x % first == 0) and (x % second == 0):
      return x
    x += 1

def gcd_nums(nums):
  for y in range(1, len(nums) - 1):
    nums[y] = gcd(nums[y-1], nums[y])
  return nums[y]

def lcm_nums(nums):
  for y in range(1, len(nums)):
    nums[y] = lcm(nums[y-1], nums[y])
  return nums[y]
```

```python
nums = input('Введите чиcла через пробел: ').split()
nums = list(map(int, nums))
nums1 = nums.copy()
gcd_d = gcd_nums(nums)
lcm_d = lcm_nums(nums1)

print('НОД = ', gcd_d)
print('НОК = ', lcm_d)

# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```

```python
def symbol_statistics(text):
  stat={}
  for letter in text:
    letter=letter.lower()
    stat[letter]=stat.get(letter, 0)+1
  return stat

text = input('Введите предложение: ')
stat = symbol_statistics(text)
for symbol in sorted(stat):
  print(symbol, '=', stat[symbol])
# --------------
# Пример вывода:
#
# Введите предложение: мама МЫла РамУ
# {'л': 1, 'р': 1, 'у': 1, 'м': 4, 'а': 4, 'ы': 1, ' ': 2}
```

#№ 5.2.14

Напишите функцию, которая принимает неограниченное количество числовых аргументов и возвращает кортеж из двух списков:

• отрицательных значений (отсортирован по убыванию);

• неотрицательных значений (отсортирован по возрастанию).

```python
def split_numbers(*args):
  positive = []
  negative = []
  for x in args:
    if x < 0:
      negative.append(x)
      negative_sort = sorted(negative)
      negative_sort.reverse()
    else:
      positive.append(x)
      positive_sort = sorted(positive)
  return negative_sort, positive_sort
```

```python
print(split_numbers(1, 4, -5, 0, -33))
# --------------
# Пример вывода:
#
# ([-5, -33], [0, 1, 4])
```

Составьте две функции для возведения числа в степень: один из вариантов реализуйте в рекурсивном стиле.

```python
def pow1(value,base):
  if base == 0: 
    return 1
  else:
    return value**base

def recursion(value, base, result=1):
    if base == 0:
        return result
    else:
        result *= value
        return recursion(value, base - 1, result)
```

```python
print(pow1(5,3))
print(recursion(5,3))
# --------------
# Пример вывода:
#
# 125
# 125
```

#№ 5.2.17

Дано натуральное число. Напишите рекурсивные функции для определения:

• суммы цифр числа;

• количества цифр в числе.

```python
def digits_sum(value):
  if value == 0:
    return 0
  else:
    return (value%10) + digits_sum(value//10) 

def digits_count(value):
  if value == 1:
    return 1
  else:
    return value**0 + digits_count(value//10)
```

```python
print(digits_sum(12345))
print(digits_count(12345))
# --------------
# Пример вывода:
#
# 15
# 5
```
